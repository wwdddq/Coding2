```python
import urllib.request
import re

# Read the source code of a web page
url = "http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing="
nothing = "6175.2" # First value of nothing
pattern = re.compile("and the next nothing is (\d+)") # Regular expression to match the next nothing

# Loop until the last nothing is found
while True:
    response = urllib.request.urlopen(url + nothing)
    data = response.read().decode()
    print(data)
    match = pattern.search(data)
    if match is None:
        break
    nothing = match.group(1)
```
